## Data loading

### Subtask:
Load the data from "combined_data.csv" into a dataframe.


**Reasoning**:
Import pandas and load the data from "combined_data.csv" into a dataframe, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv('combined_data.csv')
display(df.head())
```

## Data exploration

### Subtask:
Explore the loaded data to understand its structure, columns, data types, and basic statistics. This will help identify potential areas for cleaning and analysis.


**Reasoning**:
Explore the dataframe by displaying its information, head, descriptive statistics, missing values and shape.



```python
display(df.info())
display(df.head())
display(df.describe())
display(df.isnull().sum())
display(df.shape)
```

## Data cleaning

### Subtask:
Clean the data by handling missing values, removing duplicates, and addressing any inconsistencies found during exploration.


**Reasoning**:
Based on the previous exploration, there are no missing values. I will check for duplicate rows and remove them if found. Then I will display the number of missing values again and the shape of the dataframe to confirm.



```python
df.drop_duplicates(inplace=True)
display(df.isnull().sum())
display(df.shape)
```

## Data analysis

### Subtask:
Perform comprehensive data analysis on the cleaned data to derive insights. This may include calculating descriptive statistics, identifying trends, and exploring relationships between variables.


**Reasoning**:
Calculate and display descriptive statistics for numerical columns, the correlation matrix, and unique values/counts for categorical columns to understand the data's characteristics and potential relationships between variables.



```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
display("Descriptive Statistics:")
display(df[numerical_cols].describe())

display("\nCorrelation Matrix:")
display(df[numerical_cols].corr())

categorical_cols = df.select_dtypes(include=['object']).columns
display("\nUnique Values and Counts for Categorical Columns:")
for col in categorical_cols:
    display(f"\n--- {col} ---")
    display(df[col].value_counts())
```

## Data visualization

### Subtask:
Generate colorful charts and graphs based on the data analysis results to visually represent key findings and patterns.


**Reasoning**:
Generate colorful charts and graphs based on the data analysis results to visually represent key findings and patterns. The instructions require creating a heatmap of the correlation matrix, bar charts or count plots for 'skenario', scatter plots for highly correlated numerical variables, and potentially line plots for trends over 'tahun'. I will start by creating the heatmap, followed by the bar chart for 'skenario', and then scatter plots for highly correlated variables.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style('whitegrid')

# 1. Heatmap of the correlation matrix
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()

# 2. Bar chart/Count plot for 'skenario'
plt.figure(figsize=(8, 6))
sns.countplot(x='skenario', data=df, palette='viridis')
plt.title('Distribution of Skenario')
plt.xlabel('Skenario')
plt.ylabel('Count')
plt.show()

# 3. Scatter plots for highly correlated numerical variables
# From the correlation matrix, 'tahun', 'pendapatan', 'beban_operasional', and 'penyusutan' show high positive correlations.
# Let's visualize some of these relationships.

# Scatter plot: 'tahun' vs 'pendapatan'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='tahun', y='pendapatan', data=df, hue='skenario', s=100, palette='viridis')
plt.title('Tahun vs Pendapatan')
plt.xlabel('Tahun')
plt.ylabel('Pendapatan')
plt.show()

# Scatter plot: 'pendapatan' vs 'beban_operasional'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pendapatan', y='beban_operasional', data=df, hue='skenario', s=100, palette='viridis')
plt.title('Pendapatan vs Beban Operasional')
plt.xlabel('Pendapatan')
plt.ylabel('Beban Operasional')
plt.show()

# Scatter plot: 'pendapatan' vs 'penyusutan'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pendapatan', y='penyusutan', data=df, hue='skenario', s=100, palette='viridis')
plt.title('Pendapatan vs Penyusutan')
plt.xlabel('Pendapatan')
plt.ylabel('Penyusutan')
plt.show()

# Scatter plot: 'beban_operasional' vs 'penyusutan'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='beban_operasional', y='penyusutan', data=df, hue='skenario', s=100, palette='viridis')
plt.title('Beban Operasional vs Penyusutan')
plt.xlabel('Beban Operasional')
plt.ylabel('Penyusutan')
plt.show()

# 4. Line plots to show trends over 'tahun' for relevant numerical variables
# 'pendapatan', 'beban_operasional', and 'penyusutan' show clear trends over 'tahun'.
# Let's plot these trends, potentially separated by 'skenario'.

plt.figure(figsize=(12, 6))
sns.lineplot(x='tahun', y='pendapatan', hue='skenario', data=df, marker='o')
plt.title('Trend of Pendapatan over Tahun by Skenario')
plt.xlabel('Tahun')
plt.ylabel('Pendapatan')
plt.xticks(df['tahun'].unique())
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='tahun', y='beban_operasional', hue='skenario', data=df, marker='o')
plt.title('Trend of Beban Operasional over Tahun by Skenario')
plt.xlabel('Tahun')
plt.ylabel('Beban Operasional')
plt.xticks(df['tahun'].unique())
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='tahun', y='penyusutan', hue='skenario', data=df, marker='o')
plt.title('Trend of Penyusutan over Tahun by Skenario')
plt.xlabel('Tahun')
plt.ylabel('Penyusutan')
plt.xticks(df['tahun'].unique())
plt.show()
```
